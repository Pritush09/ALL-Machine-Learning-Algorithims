```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
data  = load_digits()
```

```python
X = data.data
y = data.target
```

```python
X.shape
```

```python
plt.imshow(X[2].reshape(8,8))
```

```python
X_train , X_test , y_train , y_test  = train_test_split(X,y,test_size=0.15)
```

```python
rf = RandomForestClassifier(n_estimators=2)
```

```python
rf.fit(X_train,y_train)
```

```python
rf.estimators_[0].feature_importances_
```

```python
rf.estimators_[1].feature_importances_
```

```python
rf.feature_importances_
```

```python
plt.figure(figsize=(15,9))
sns.heatmap(rf.feature_importances_.reshape(8,8),annot=True)
```

* we can clearly see that the side lines are not that important in the prediction of the target column

```python
dt = DecisionTreeClassifier()
dt.fit(X_train,y_train)
```

```python
dt.feature_importances_
```
